### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
'''
Indicaciones TRELLO
Ingreso total por cada tienda
En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.
'''
# Ingreso Total por cada Tienda

try:
  print(f"Ingreso Total Tienda 1 es: {sum(tienda['Precio'])}")
  print(f"Ingreso Total Tienda 2 es: {sum(tienda2['Precio'])}")
  print(f"Ingreso Total Tienda 3 es: {sum(tienda3['Precio'])}")
  print(f"Ingreso Total Tienda 4 es: {sum(tienda4['Precio'])}")
except:
  print("Error")

```

# 2. Ventas por categoría

```python
'''
Indicaciones TRELLO
Ventas por categoría
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.
'''
#Ventas por categoría TIENDA 1
try:
  conteo = tienda['Categoría del Producto'].value_counts()
  conteo = conteo.reset_index(name='Cantidad Vendida')
  print(f'Las cantidad de ventas por Categoría en TIENDA 1 es:\n{conteo}')
  print(f'Total Cantidad de Ventas en Tienda 1:{len(tienda)}\n\n\n')
except:
  print("Error")

#Ventas por categoría TIENDA 2
try:
  conteo2 = tienda2['Categoría del Producto'].value_counts()
  conteo2 = conteo2.reset_index(name='Cantidad Vendida')
  print(f'Las cantidad de ventas por Categoría en TIENDA 2 es:\n{conteo2}')
  print(f'Total Cantidad de Ventas en Tienda 2:{len(tienda2)} \n\n\n')
except:
  print("Error")

#Ventas por categoría TIENDA 3
try:
  conteo3 = tienda3['Categoría del Producto'].value_counts()
  conteo3 = conteo3.reset_index(name='Cantidad Vendida')
  print(f'Las cantidad de ventas por Categoría en TIENDA 3 es:\n{conteo3}')
  print(f'Total Cantidad de Ventas en Tienda 3:{len(tienda3)}\n\n\n')
except:
  print("Error")

#Ventas por categoría TIENDA 4
try:
  conteo4 = tienda4['Categoría del Producto'].value_counts()
  conteo4 = conteo4.reset_index(name='Cantidad Vendida')
  print(f'Las cantidad de ventas por Categoría en TIENDA 4 es:\n{conteo4}')
  print(f'Total Cantidad de Ventas en Tienda 4:{len(tienda4)}\n\n\n')
except:
  print("Error")

```

# 3. Calificación promedio de la tienda


```python
'''
Indicaciones TRELLO
Valoración media por tienda
En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

'''
# Valoración Media por cada Tienda

try:
  print(f"Valoración Media Tienda 1 es: {round(tienda['Calificación'].mean(),2)}")
  print(f"Valoración Media Tienda 2 es: {round(tienda2['Calificación'].mean(),2)}")
  print(f"Valoración Media Tienda 3 es: {round(tienda3['Calificación'].mean(),2)}")
  print(f"Valoración Media Tienda 4 es: {round(tienda4['Calificación'].mean(),2)}")
except:
  print("Error")

```

# 4. Productos más y menos vendidos

```python
'''
Indicaciones TRELLO
Productos más vendidos y menos vendidos
En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.
'''

#Productos más vendidos y menos vendidos TIENDA 1
try:
  conteo = tienda['Producto'].value_counts()
  max = conteo.max()
  min = conteo.min()
  conteo = conteo.reset_index(name='Cantidad Vendida')
  conteomax = conteo[conteo['Cantidad Vendida'] == max]
  conteomin = conteo[conteo['Cantidad Vendida'] == min]
  print(f'Los productos más vendido en Tienda 1 son:\n {conteomax}\n')
  print(f'Los productos menos vendido en Tienda 1 son:\n {conteomin}\n\n\n')
except:
  print("Error")

#Productos más vendidos y menos vendidos TIENDA 2
try:
  conteo2 = tienda2['Producto'].value_counts()
  max = conteo2.max()
  min = conteo2.min()
  conteo2 = conteo2.reset_index(name='Cantidad Vendida')
  conteomax = conteo2[conteo2['Cantidad Vendida'] == max]
  conteomin = conteo2[conteo2['Cantidad Vendida'] == min]
  print(f'Los productos más vendido en Tienda 2 son:\n {conteomax}\n')
  print(f'Los productos menos vendido en Tienda 2 son:\n {conteomin}\n\n\n')
except:
  print("Error")

#Productos más vendidos y menos vendidos TIENDA 3
try:
  conteo3 = tienda3['Producto'].value_counts()
  max = conteo3.max()
  min = conteo3.min()
  conteo3 = conteo3.reset_index(name='Cantidad Vendida')
  conteomax = conteo3[conteo3['Cantidad Vendida'] == max]
  conteomin = conteo3[conteo3['Cantidad Vendida'] == min]
  print(f'Los productos más vendido en Tienda 3 son:\n {conteomax}\n')
  print(f'Los productos menos vendido en Tienda 3 son:\n {conteomin}\n\n\n')
except:
  print("Error")


#Productos más vendidos y menos vendidos TIENDA 4
try:
  conteo4 = tienda4['Producto'].value_counts()
  max = conteo4.max()
  min = conteo4.min()
  conteo4 = conteo4.reset_index(name='Cantidad Vendida')
  conteomax = conteo4[conteo4['Cantidad Vendida'] == max]
  conteomin = conteo4[conteo4['Cantidad Vendida'] == min]
  print(f'Los productos más vendido en Tienda 4 son:\n {conteomax}\n')
  print(f'Los productos menos vendido en Tienda 4 son:\n {conteomin}\n\n\n')
except:
  print("Error")
```

# 5. Envío promedio por tienda

```python
'''
Indicaciones TRELLO
Valor del envío promedio por tienda
En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.
'''

# Valor del envío promedio por tienda

try:
  print(f"Valor del envío promedio Tienda 1 es: {int(tienda['Costo de envío'].mean())}")
  print(f"Valor del envío promedio Tienda 2 es: {int(tienda2['Costo de envío'].mean())}")
  print(f"Valor del envío promedio Tienda 3 es: {int(tienda3['Costo de envío'].mean())}")
  print(f"Valor del envío promedio Tienda 4 es: {int(tienda4['Costo de envío'].mean())}")
except:
  print("Error")

```

```python
'''
Indicaciones TRELLO
Generando gráfico
Después de realizar los análisis, es hora de transformar sus resultados en visualizaciones que le ayuden a comprender mejor los patrones y los insights encontrados.

De acuerdo con funciones que haya creado, le recomendamos que genere al menos tres gráficos. Estos gráficos deben ser de diferentes tipos (como barras, líneas, dispersión, entre otros) para presentar una visión completa de los datos.

Recuerde que los gráficos deben complementar el análisis realizado, resaltando los puntos más relevantes, como los ingresos de la tienda, la distribución de categorías de productos, las opiniones de los clientes, los productos más vendidos y/o los costes de envío.

No es necesario mostrar una visualización en cada paso, pero al final del análisis debes generar los gráficos que consideres más útiles para interpretar los datos.

Consejo: Matplotlib ofrece una variedad de tipos de gráficos que son fáciles de implementar, lo que lo convierte en una excelente opción para principiantes. Utilice su creatividad para elegir los gráficos que mejor resalten las tendencias y los conocimientos de su análisis.
'''

import matplotlib.pyplot as grafico
# Datos para el gráfico
valores = [sum(tienda['Precio']), sum(tienda2['Precio']), sum(tienda3['Precio']), sum(tienda4['Precio'])]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear el gráfico
grafico.pie(valores, labels=nombres, autopct='%1.1f%%', startangle=90)

# Añadir un título
grafico.title('Gráfico Ingreso Total por cada Tienda')

# Mostrar el gráfico
grafico.show()


```
